```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 100, 50)
plt.plot(x, 2*x)
plt.show()
plt.plot(2*x)
plt.plot(3*x)
plt.show()
```

```python
x = np.linspace(0, 2 * np.pi, 16)
plt.subplot(2, 2, 1)   # (row, column, activatedNo.)
plt.plot(x, np.sin(x), ':*r') # 线性，点性，颜色
plt.subplot(2, 2, 4)
plt.plot(x, np.cos(x), 'b')
plt.show()
```

```python
x = np.linspace(0, 2 * np.pi, 20)
plt.plot(x, np.sin(x), label='Sin Function')
plt.legend()
plt.xlabel('x label')
plt.ylabel('y label')
plt.show()
```
